```python
import cv2
import numpy as np

# Load the first image
img1 = cv2.imread('flower.jpeg')

# Load the second image
img2 = cv2.imread('onion.png')

# Ensure both images have the same size
img2_resized = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

# Create an empty array for the result
subtracted_image = np.zeros_like(img1, dtype=np.uint8)

# Subtract the images pixel-wise
for i in range(img1.shape[0]):
    for j in range(img1.shape[1]):
        subtracted_image[i, j] = np.clip(img1[i, j].astype(int) - img2_resized[i, j].astype(int), 0, 255).astype(np.uint8)

# Display the result
cv2.imshow('Subtracted Image', subtracted_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Load the first image
img1 = cv2.imread('flower.jpeg')

# Load and resize the second image to match the dimensions of img1
img2 = cv2.imread('onion.png')
img_re = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

# Ensure both images have the same data type
img_re = img_re.astype(img1.dtype)

# Add the images
added_image = cv2.add(img1, img_re)

# Display the result
cv2.imshow('Added Image', added_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Load the first image
img1 = cv2.imread('flower.jpeg')

# Load the second image
img2 = cv2.imread('onion.png')

# Ensure both images have the same size
img2_resized = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

# Create an empty array for the result
added_image = np.zeros_like(img1, dtype=np.uint8)

# Subtract the images pixel-wise
for i in range(img1.shape[0]):
    for j in range(img1.shape[1]):
        added_image[i, j] = np.clip(img1[i, j].astype(int) + img2_resized[i, j].astype(int), 0, 255).astype(np.uint8)

# Display the result
cv2.imshow('added Image', added_image) 
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image
I = cv2.imread('bitplane.bmp', cv2.IMREAD_GRAYSCALE)

# Bitplane slicing
G1 = np.uint8((I & 0b00000001) * 255)  # get bitplane 1
G2 = np.uint8((I & 0b00000010) * 255)  # get bitplane 2
G3 = np.uint8((I & 0b00000100) * 255)  # get bitplane 3
G4 = np.uint8((I & 0b00001000) * 255)  # get bitplane 4
G5 = np.uint8((I & 0b00010000) * 255)  # get bitplane 5
G6 = np.uint8((I & 0b00100000) * 255)  # get bitplane 6
G7 = np.uint8((I & 0b01000000) * 255)  # get bitplane 7
G8 = np.uint8((I & 0b10000000) * 255)  # get bitplane 8

# Plotting the results
plt.figure(figsize=(10, 8))

plt.subplot(4, 4, 1)
plt.imshow(I, cmap='gray')
plt.title('Original 8 bit image')

plt.subplot(4, 4, 9)
plt.imshow(G1, cmap='gray')
plt.title('Bitplane 1')

plt.subplot(4, 4, 10)
plt.imshow(G2, cmap='gray')
plt.title('Bitplane 2')

plt.subplot(4, 4, 11)
plt.imshow(G3, cmap='gray')
plt.title('Bitplane 3')

plt.subplot(4, 4, 12)
plt.imshow(G4, cmap='gray')
plt.title('Bitplane 4')

plt.subplot(4, 4, 13)
plt.imshow(G5, cmap='gray')
plt.title('Bitplane 5')

plt.subplot(4, 4, 14)
plt.imshow(G6, cmap='gray')
plt.title('Bitplane 6')

plt.subplot(4, 4, 15)
plt.imshow(G7, cmap='gray')
plt.title('Bitplane 7')

plt.subplot(4, 4, 16)
plt.imshow(G8, cmap='gray')
plt.title('Bitplane 8')
plt.show()
```
