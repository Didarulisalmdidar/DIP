# Ideal Low Pass Filter

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the image
f = cv2.imread('cat.jpeg', 0)  # Load the image in grayscale mode 

# Perform Fourier transform
f_transform = np.fft.fft2(f) 
f_shift = np.fft.fftshift(f_transform)
plt.imshow(f,cmap='gray')
plt.title('Original Image')
plt.show() 
plt.imshow(np.abs(f_transform),cmap='gray')
plt.title('Centered Spectrum')
#plt.axis('')
plt.show()
plt.imshow(np.abs(f_shift),cmap='gray')
plt.title('Fourier Spectrum')
plt.show() 
plt.imshow(np.log(np.abs(f_shift)),cmap='gray')
plt.title('Enhanced spectrum')
plt.show() 

# Filter: Low pass filter
M,N = f.shape
H = np.zeros((M,N), dtype=np.float32)
D0 = 50
for u in range(M):
    for v in range(N):
        D = np.sqrt((u-M/2)**2 + (v-N/2)**2)
        if D <= D0:
            H[u,v] = 1
        else:
            H[u,v] = 0

plt.imshow(np.abs(H),cmap='gray')
plt.title('Low pass filter')
plt.show()  

# Apply the filter H
f_filtered = f_shift * H
plt.imshow(np.log1p(np.abs(f_filtered)),cmap='gray')
plt.title('Applying filter')
plt.show()

# Perform inverse Fourier transform
f_ishift=np.fft.ifftshift(f_filtered)
plt.imshow(np.log1p(np.abs(f_ishift)),cmap='gray')
plt.title('Inverse Transform')
plt.show() 
image_filtered = np.fft.ifft2(f_ishift)
# Take the absolute value
image_filtered = np.abs(image_filtered)
plt.imshow(image_filtered,cmap='gray')
plt.title('Filtered Image ')
plt.show() 

```

# Gaussian Low Pass Filter

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the image
f = cv2.imread('cat.jpeg', 0)  # Load the image in grayscale mode 

# Perform Fourier transform
f_transform = np.fft.fft2(f) 
f_shift = np.fft.fftshift(f_transform)
plt.imshow(f,cmap='gray')
plt.title('Original Image')
plt.show() 
plt.imshow(np.abs(f_transform),cmap='gray')
plt.title('Centered Spectrum')
#plt.axis('')
plt.show()
plt.imshow(np.abs(f_shift),cmap='gray')
plt.title('Fourier Spectrum')
plt.show() 
plt.imshow(np.log(np.abs(f_shift)),cmap='gray')
plt.title('Enhanced spectrum')
plt.show() 

# Filter: Gaussian Low pass filter
M,N = f.shape
H = np.zeros((M,N), dtype=np.float32)
Do = 10
for u in range(M):
    for v in range(N):
        D = np.sqrt((u-M/2)**2 + (v-N/2)**2)
        H[u,v]=np.exp(-D**2/(2*Do*Do))

plt.imshow(np.abs(H),cmap='gray')
plt.title(' Gaussian Low pass filter')
plt.show()  

# Apply the filter H
f_filtered = f_shift * H
plt.imshow(np.log1p(np.abs(f_filtered)),cmap='gray')
plt.title('Applying filter')
plt.show()

# Perform inverse Fourier transform
f_ishift=np.fft.ifftshift(f_filtered)
plt.imshow(np.log1p(np.abs(f_ishift)),cmap='gray')
plt.title('Inverse Transform')
plt.show() 
image_filtered = np.fft.ifft2(f_ishift)
# Take the absolute value
image_filtered = np.abs(image_filtered)
plt.imshow(image_filtered,cmap='gray')
plt.title('Filtered Image ')
plt.show() 

```

# Butter worth Low Pass Filter

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the image
f = cv2.imread('cat.jpeg', 0)  # Load the image in grayscale mode 

# Perform Fourier transform
f_transform = np.fft.fft2(f) 
f_shift = np.fft.fftshift(f_transform)
plt.imshow(f,cmap='gray')
plt.title('Original Image')
plt.show() 
plt.imshow(np.abs(f_transform),cmap='gray')
plt.title('Centered Spectrum')
#plt.axis('')
plt.show()
plt.imshow(np.abs(f_shift),cmap='gray')
plt.title('Fourier Spectrum')
plt.show() 
plt.imshow(np.log(np.abs(f_shift)),cmap='gray')
plt.title('Enhanced spectrum')
plt.show() 

# Filter: Gaussian Low pass filter
M,N = f.shape
H = np.zeros((M,N), dtype=np.float32)
Do = 10
n=5 # filter order
for u in range(M):
    for v in range(N):
        D = np.sqrt((u-M/2)**2 + (v-N/2)**2)
        H[u,v]=1/(1+(D/Do)**n) 

plt.imshow(np.abs(H),cmap='gray')
plt.title(' Butter Low pass filter')
plt.show()  

# Apply the filter H
f_filtered = f_shift * H
plt.imshow(np.log1p(np.abs(f_filtered)),cmap='gray')
plt.title('Applying filter')
plt.show()

# Perform inverse Fourier transform
f_ishift=np.fft.ifftshift(f_filtered)
plt.imshow(np.log1p(np.abs(f_ishift)),cmap='gray')
plt.title('Inverse Transform')
plt.show() 
image_filtered = np.fft.ifft2(f_ishift)
# Take the absolute value
image_filtered = np.abs(image_filtered)
plt.imshow(image_filtered,cmap='gray')
plt.title('Filtered Image ')
plt.show() 

```
