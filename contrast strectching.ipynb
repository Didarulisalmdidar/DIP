```python
import pandas as pd 
import numpy as np 
import cv2
import matplotlib.pyplot as plt  
```

```python
r=cv2.imread('bitplane.bmp',cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(r,cmap='gray')
```

```python
# now take input r1,r2,s1,s2
r1=float(input('Enter r1: '))
r2=float(input('Enter r2: '))
s1=float(input('Enter s1: '))
s2=float(input('Enter s2: '))

# calculate slop
a=s1/r1
b=(s2-s1)/(r2-r1)
c=(255-s2)/(255-r2)
s=np.zeros_like(r) 
for i in range(r.shape[0]):
    for j in range(r.shape[1]):
        if r[i][j]<r1:
            s[i][j]=a*r[i][j]
        elif r[i][j]<r2:
            s[i][j]=b*(r[i][j]-r1)+s1
        else:
            s[i][j]=c*(r[i][j]-r2)+s2
            
plt.figure() 
plt.subplot(1,2,1)
plt.imshow(r,cmap='gray') 
plt.title('Original')
plt.subplot(1,2,2) 
plt.imshow(s,cmap='gray')
plt.title('Contrast Stretching')
```

```python
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
plt.hist(r.ravel(),bins=256,range=[0,256]) 
plt.subplot(1,2,2)
plt.hist(s.ravel(),bins=256,range=[0,256])
plt.show() 
```
