```python
#2d dft

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image in grayscale
image = cv2.imread('cat.jpeg', cv2.IMREAD_GRAYSCALE)

# Compute the 2D DFT
dft = np.fft.fft2(image)

# Shift the zero frequency component to the center
dft_shift = np.fft.fftshift(dft)

# Compute the magnitude spectrum
magnitude_spectrum = np.abs(dft_shift)

# Enhance spectrum by logarithmic transformation
enhanced_spectrum = np.log(magnitude_spectrum + 1)

# Plot the images
plt.figure(figsize=(10, 8))

# Original Image
plt.subplot(221)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Fourier Spectrum
plt.subplot(222)
plt.imshow(magnitude_spectrum, cmap='gray')
plt.title('Fourier Spectrum')
plt.axis('off')

# Centered Spectrum
plt.subplot(223)
plt.imshow(np.abs(dft), cmap='gray')
plt.title('Centered Spectrum')
plt.axis('off')

# Enhanced Spectrum (Log Transformation)
plt.subplot(224)
plt.imshow(enhanced_spectrum, cmap='gray')
plt.title('Enhanced Spectrum (Log Transform)')
plt.axis('off')

plt.tight_layout()
plt.show()
#2D DFT 
```

```python
I=cv2.imread('cat.jpeg',cv2.IMREAD_GRAYSCALE)
h = (1/9) * np.ones((3,3), dtype=np.float32) # filter

# Perform spatial filtering (convolution)
I2 = cv2.filter2D(I,-1, h)

# Display the original and filtered images
cv2.imshow('Original Image', I)
cv2.imshow('Filtered Image', I2)
cv2.waitKey(0)
cv2.destroyAllWindows()
# filter image got some blurred
```
